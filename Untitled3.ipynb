```python
import pandas as pd
df = pd.DataFrame({'C':[1,1,1,2,2,2],'X':['c','a','b','c','c','b',],'Y':['b','d','d','f','f','f']})
df
```

```python
grouped = df.groupby('C')[['X','Y']].agg(['unique'])
grouped
```

```python
grouped.columns = [column[0] for column in grouped.columns]
grouped.reset_index()
```

```python
grouped['Total number of X'] = grouped.X.map(len)
grouped['Total number of Y'] = grouped.Y.map(len)
grouped
```

```python
grouped['X joined'] = grouped.Y.map(','.join)
grouped
```
